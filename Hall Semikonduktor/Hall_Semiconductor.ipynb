```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline
```

# Fungsi Arus

## Tipe N-Ge (B = 182,9 mT)

```python
d_NI1 = pd.read_csv('./Fungsi Arus Tipe N (B = 182,9 mT).txt', sep = '\t')

x = d_NI1["Current I_A1 / A"]
y = d_NI1["Voltage U_B1 / V"]
x_const = sm.add_constant(x)
hasil = sm.OLS(y, x_const).fit()

const = hasil.params[0]
grad = hasil.params[1]
print("Persamaan:\ny =", grad, "x +", const)

grad_err = hasil.bse[1]
const_err = hasil.bse[0]
print("Kesalahan: \ndelta m = ", grad_err, "\ndelta c = ", const_err)

y_predict = grad*x + const
plt.plot(x, y, label = 'percobaan')
plt.plot(x, y_predict, label = 'regresi')
plt.legend()
plt.xlabel("I (A)")
plt.ylabel("U (V)")
plt.title("Tipe N-Ge, B = 182,9 mT")
plt.grid()
plt.show()

print(hasil.summary())
```
