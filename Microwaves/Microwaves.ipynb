```python
import pandas as pd
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

## Percobaan 1

```python
P1 = pd.read_csv('Percobaan 1.txt','\t')
print(P1)

y = P1["Y(mm)"]
x1 = P1["X100 mm (V)"]
x2 = P1["X200 mm (V)"]

plt.plot(y, x1, label = 'X0 = 100mm', marker = 'o')
plt.plot(y, x2, label = 'X0 = 200mm', marker = 'o')
plt.grid()
plt.xlabel('Jarak (mm)')
plt.ylabel('Tegangan (V)')
plt.title("Grafik Percobaan Distribusi Bidang Transversal")
plt.legend()
plt.show()
```

## Percobaan 2

```python
P2 = pd.read_csv('Percobaan 2.txt','\t')
print(P2)

y = P2["U/V (V)"]
x = P2["theta"]

plt.plot(x, y, marker = 'o')
plt.grid()
plt.xlabel('Sudut polarisasi (grd)')
plt.ylabel('Tegangan (V)')
plt.title("Grafik Percobaan Polarisasi")
plt.show()
```
