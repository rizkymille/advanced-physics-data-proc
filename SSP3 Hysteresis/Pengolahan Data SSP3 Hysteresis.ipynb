```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

filename = ["v1.5", "v1", "v2.5", "v2", "v3"]
```

# Bahan 1

```python
for data in filename:
    df = pd.read_csv(f"Data Percobaan/1/{data}.txt", sep="\t", skiprows=4)
    idx_magflux_max = df["Magnetic flux &F / Vs"].idxmax()
    df = df[:idx_magflux_max*6]
    plt.plot(df["Current I_A1 / A"], df["Magnetic flux &F / Vs"])
    plt.title(f"{data}")
    plt.show()
    df_obs = df[["Current I_A1 / A", "Magnetic flux &F / Vs"]]
    df_obs.to_csv(f"Data Pengamatan/1/{data}.txt", sep="\t", index=False)
```
