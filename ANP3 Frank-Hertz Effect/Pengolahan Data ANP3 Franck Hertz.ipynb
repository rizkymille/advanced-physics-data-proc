```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline
```

# Grafik A

```python
df_graphA = pd.read_csv("Data Grafik A.txt", sep="\t")
plt.plot(df_graphA["U (V)"], df_graphA["I (nA)"])
plt.title("Grafik A")
plt.xlabel("U2 (V)")
plt.ylabel("I (nA)")
plt.show()
```

## Regresi linier

```python
df_peakA = pd.read_csv("Peak Grafik A.txt", sep="\t")
x_peakA = df_peakA["n"]
y_peakA = df_peakA["U2 (V)"]

addConst_peakA = sm.add_constant(x_peakA)
result_peakA = sm.OLS(y_peakA, addConst_peakA).fit()

const_peakA = result_peakA.params[0]
grad_peakA = result_peakA.params[1]
print(f"Persamaan:\ny = {grad_peakA} x + {const_peakA}")

gradErr_peakA = result_peakA.bse[1]
print("Kesalahan: \ndelta m =", gradErr_peakA)

yPred_peakA = grad_peakA*x_peakA+const_peakA
plt.plot(x_peakA,y_peakA,'o',x_peakA,yPred_peakA)
plt.title("Regresi linier grafik A")
plt.xlabel("n")
plt.ylabel("U2")
plt.show()
```

## Panjang gelombang dan Kesalahan relatif

```python
h = 6.626 * 1e-34
c = 299792458
q = 1.502 * 1e-19

wavelen_A = (h*c)/(q*grad_peakA)
print("Panjang gelombang =", wavelen_A)

wavelenErr_A = (h*c*gradErr_peakA)/(q*(grad_peakA)**2)
print("Kesalahan panjang gelombang =", wavelenErr_A)

relErr_A = abs(wavelenErr_A/wavelen_A)*100
print(f"Kesalahan relatif = {relErr_A}%")
```
